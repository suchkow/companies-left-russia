Below I've reflected the decoding of abbreviations in a nice small table.

```python
regs_ls = [
    'NE','EE','WE','SE',
    'NAf','EAf','WAf','SAf','CAf',
    'AO',
    'NAs','EAs','WAs','SAs','CAs','SEAs',
    'NAm','SAm','CAm',
    'C'
]

regs_full_ls = [
    'Northern Europe', 'Western Europe', 'Eastern Europe', 'Southern Europe',
    'North Africa', 'West Africa', 'East Africa', 'South Africa', 'Central Africa',
    'Australia and Oceania',
    'North Asia', 'Western Asia', 'East Asia', 'South Asia', 'Central Asia', 'Southeast Asia',
    'North America', 'South America', 'Central America',
    'Caribbean'
]

regs_abb_df = pd.DataFrame(
    list(zip(regs_ls, regs_full_ls)), columns =['Abbreviation', 'Region']
)

regs_abb_df.head(20)
```
